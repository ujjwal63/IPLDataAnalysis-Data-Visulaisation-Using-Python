```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

### LOAD DATA

```python
data= pd.read_csv("./IPL Matches 2008-2020.csv")
```

```python
data.head()
```

```python
data.info()
```

```python
data.drop(['method'], axis=1, inplace=True)
```

```python
data.info()
```

### MOST WINS IN IPL 

```python
t= pd.DataFrame({"Winner": data['winner']})
wins= t.value_counts()
labels= [X[0] for X in wins.keys()]
bar,ax=plt.subplots(figsize=(20,12))
ax=plt.pie(x=wins, autopct="%.1f%%", labels=labels)
plt.title("Most Wins in IPL", fontsize=15)
plt.show()
```

### MOST WINS IN ELIMINATOR

```python
sns.countplot(data['winner'][data['eliminator']=='Y'], data=data)
plt.title("Most Wins in Eliminator",fontsize=15)
plt.xticks(rotation=90)
plt.show()
```

### TOSS DECISION

```python
teams= data['toss_winner'].unique()
decision= pd.DataFrame([], columns=["Toss Winner", "Decision", "Times"])

for id,ele in enumerate(teams):
    bats = data[(data['toss_winner']==ele) & (data['toss_decision']== "bat")]
    fields = data[(data['toss_winner']==ele) & (data['toss_decision']== "field")]
    decision=decision.append({"Toss Winner": ele, "Decision": "bat", "Times": bats['toss_winner'].count()}, ignore_index=True)
    decision=decision.append({"Toss Winner": ele, "Decision": "field", "Times": fields['toss_winner'].count()}, ignore_index=True)

decision
```

```python
sns.catplot(x="Toss Winner", y="Times", hue="Decision", data=decision, kind="bar", height=5, aspect=2)
plt.xticks(rotation=90)
plt.title("Toss Decision of Teams")
plt.xlabel("IPL Teams")
plt.ylabel("Toss Decision")
plt.show()
```
